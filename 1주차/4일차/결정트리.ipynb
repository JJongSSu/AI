```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer["data"], cancer["target"], random_state=0)
```

```python
model = DecisionTreeClassifier(random_state=0)
model.fit(X_train, y_train)
```

```python
# 과대적합 확인
print(model.score(X_train, y_train))
print(model.score(X_test, y_test))
```

```python
model = DecisionTreeClassifier(max_depth=8, max_leaf_nodes=15, random_state=0)
model.fit(X_train, y_train)
```

```python
print(model.score(X_train, y_train))
print(model.score(X_test, y_test))
```

##### data가 대부분 명목척도로 되어 있으면 확률모델을 사용하는 것이 좋음


```python
list_a = []
for a, i in enumerate(mushrooms["cap-shape"]):
    i = list(mushrooms["cap-shape"].unique()).index(i)
    list_a.append(i)
mushrooms["cap-shape"] = list_a
mushrooms["cap-shape"]
```

```python
def encode(series):
    dic = dict(zip(series.unique(), range(len(series.unique()))))
    result = series.apply(lambda x : dic[x])
    return result
```

```python
for i in mushrooms.columns:
    mushrooms[i] = encode(mushrooms[i])
mushrooms
```

```python
mushrooms.corr()
```

```python
mushrooms.corrwith(mushrooms["class"]).sort_values()
```

```python
y = mushrooms["class"]
X = mushrooms
del X["class"]
del X["veil-type"]
# del X["odor"]
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
print(X_train)
print(X_test)
print(y_train)
print(y_test)
```

```python
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
from sklearn.tree import export_graphviz


export_graphviz(model, out_file='mushroom_tree.dot',
                class_names = ['독','식용'],
                feature_names = X.columns,
                impurity = True,
                filled = True)
```
