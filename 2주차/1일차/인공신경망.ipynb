### hyper parameter
    hidden_layer_sizes 
        은닉층 크기 정의, 은닉층의 수와 뉴런의 갯수를 튜플로 전달, 기본값 = (100,)
    
    활성화 함수
        activation
            sigmoid -> logistic
            기본값 = relu
    
    alpha
        L2규제 적용, 기본값 = 0.0001
        sklearn은 L2규제만 지원
    
    batch_size : 배치크기
    learning_rate_init : 학습률 초기값 지정
    learning_rate : 학습률 크기
    max_iter : 학습 반복 횟수

```python
import pandas as pd
import numpy as np
import mglearn
import graphviz
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer, make_moons
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
```

```python
# 인공신경망에 two_moons data set 적용
# 은닛 유닛의 갯수는 따로 설정 X, 기본값 100 사용
x, y = make_moons(n_samples=100, noise=0.25, random_state=0)
x_train, x_test, y_train, y_test = train_test_split(x, y, stratify=y, random_state=13)
```

```python
mlp = MLPClassifier(random_state=0)
mlp.fit(x_train, y_train)
```

```python
mglearn.plots.plot_2d_separator(mlp, x_train, fill=True, alpha=0.3)
mglearn.discrete_scatter(x_train[:,0], x_train[:,1], y_train)
plt.xlabel('property 0')
plt.ylabel('property 1')
plt.show()
```

```python
mlp = MLPClassifier(random_state=0, hidden_layer_sizes=(10,))
mlp.fit(x_train, y_train)
mglearn.plots.plot_2d_separator(mlp, x_train, fill=True, alpha=0.3)
mglearn.discrete_scatter(x_train[:,0], x_train[:,1], y_train)
plt.xlabel('property 0')
plt.ylabel('property 1')
plt.show()
```

    은닉 유닛 추가
    은닉층 추가
    활성화 함수

```python
mlp = MLPClassifier(random_state=0, hidden_layer_sizes=(10,10))
mlp.fit(x_train, y_train)
mglearn.plots.plot_2d_separator(mlp, x_train, fill=True, alpha=0.3)
mglearn.discrete_scatter(x_train[:,0], x_train[:,1], y_train)
plt.xlabel('property 0')
plt.ylabel('property 1')
plt.show()
```

```python
mlp = MLPClassifier(random_state=0, hidden_layer_sizes=(10,10), activation='tanh')
mlp.fit(x_train, y_train)
mglearn.plots.plot_2d_separator(mlp, x_train, fill=True, alpha=0.3)
mglearn.discrete_scatter(x_train[:,0], x_train[:,1], y_train)
plt.xlabel('property 0')
plt.ylabel('property 1')
plt.show()
```
