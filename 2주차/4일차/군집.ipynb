```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage     # 시각화
from sklearn.cluster import AgglomerativeClustering     # 시각화 못함
```

```python
customer_data = pd.read_csv('shopping-data.csv')
customer_data.head()
```

```python
data = customer_data.iloc[:,3:5]
data.head()
```

```python
plt.figure(figsize=(10,10))
plt.title('customer dend')
dend = dendrogram(linkage(data, method='ward'))
# d1 = linkage(x, method='ward')
# dr1 = dendrogram(d1, orientation='left')
```

```python
fig = plt.figure(figsize=(20,10))

ax1 = fig.add_subplot(1,3,1)
d1 = dendrogram(linkage(data, method='ward'), orientation='left')

ax2 = fig.add_subplot(1,3,2)
d2 = dendrogram(linkage(data, method='ward'), orientation='right')

ax1 = fig.add_subplot(1,3,3)
d1 = dendrogram(linkage(data, method='ward'), orientation='bottom')

plt.show()
```

- sklearn은 dendrogram처럼 시각화 할 수 없음
- 군집이 어떤 과정으로 생성이 되었는지 시각화하고 싶으면 dendrogram사용

```python
cluster = AgglomerativeClustering(n_clusters=5, linkage='ward')
cluster.fit_predict(data)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(data[:,0], data[:,1], c=cluster.labels_, cmap='rainbow')    # c : data
```
