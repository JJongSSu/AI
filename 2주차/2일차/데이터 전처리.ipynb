### 데이터 전처리

1. 결측치 처리
- 제거
- 채우기

```python
import pandas as pd
import numpy as np
```

```python
d = {'score1':[100,90,np.nan,95],'score2':[85,np.nan,20,30],'score3':[88,45,73,57],'score4':[np.nan,np.nan,np.nan,65]}
```

```python
df = pd.DataFrame(d)
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df[df.score1.isnull()]
```

```python
# 행 기준 삭제
df.dropna()
```

```python
df.dropna(axis=0, how='any')    # how='any' : 결측치가 있는 행 전부를 지움
```

```python
# 열 기준 삭제
df.dropna(axis=1)
```

```python
#특정 열 안에서만 삭제
df.dropna(subset=['score2','score4'])
```

```python
# 결측치 채우기
df.fillna(0)
```

```python
# 결측치 이전 값으로 채우기
df.fillna(method='pad')
```

```python
# 결측치 뒤의 값으로 채우기
df.fillna(method='bfill')
```

```python
# 결측치의 각 열의 평균값으로 채우기
df.fillna(df.mean())
```

```python
# 결측치의 각 열의 중간값, 최대값, 최소값으로 채우기
display(df.fillna(df.min()))
display(df.fillna(df.max()))
display(df.fillna(df.median()))
```

```python
# replace로 결측치 채우기
df.replace(to_replace=np.nan, value=0)
```

```python
# 결측치 앞,뒤의 중간값
df.interpolate(method='linear', limit_direction='forward')
```

```python
# 결측치 이전 값으로 채우기
df.interpolate(method='pad', limit_direction='forward')
```

```python
# 행 전체값이 결측치인 행을 삭제
d2 = {'score1':[100,np.nan,np.nan,95],'score2':[np.nan,np.nan,np.nan,np.nan],'score3':[88,45,np.nan,57],'score4':[np.nan,np.nan,np.nan,65]}
df2 = pd.DataFrame(d2)
df2
```

```python
df2.dropna(how='all')
```

```python
df2.dropna(axis=1,how='all')
```

```python
# 임계치 설정해서 제거 -> 적당한 임계치는 분석가가 판단해서 설정
df2.dropna(thresh=2)    # thresh : 초과
```

### 범주형 data를 원핫인코딩으로 변경
- 모든 data를 0과 1로 변환
- 컴퓨터는 이진법으로 처리하기 때문에 레이블 인코딩보다는 주로 원핫 인코딩을 사용

```python
df = pd.read_csv('mushrooms.csv')
df.head()
```

```python
df['cap-shape'].unique()
```

```python
one = pd.get_dummies(df)
one
# get_dummies의 문제점
# train data에만 있고 test data에는 없는 카테고리를 원핫인코딩된 칼럼으로 바꿔주지 못함
```

```python
# sklearn onehotencoding 사용
x = df.iloc[:,1:]
y = df['class']
```

```python
len(x), len(y)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
one = OneHotEncoder(sparse=False)
one.fit(x[['cap-shape']])
```

```python
train_cap = one.transform(x[['cap-shape']])
train_cap
```

```python
one.categories_
```

```python
o = pd.DataFrame(train_cap, columns=['cap-shape_'+ col for col in one.categories_])
o
```

```python
one_x = pd.concat([x.drop(columns=['cap-shape']), o], axis=1)
one_x.head(7)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(one_x, y, random_state=0)
```

```python
len(x_train), len(y_train), len(x_test), len(y_test)
```

### 범주특성을 레이블 인코딩으로 변환
- 숫자의 크고 작음에 대한 특성이 작용함
- 회귀와 같이 연속된 실수를 다루는 알고리즘에서는 레이블 인코딩을 사용하면 잘못된 결과값이 나올 수 있음

```python
one_x['cap-surface'].unique()
```

```python
one_x['cap-surface'].value_counts()
```

```python
# 인코딩 전
one_x['cap-surface']
```

```python
# 인코딩 후 - map 사용
one_x['cap-surface'] = one_x['cap-surface'].map({'y':0,'s':1,'f':2,'g':3})
one_x['cap-surface']
```

```python
one_x['cap-surface'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
encoder.fit(one_x['cap-color'])
```

```python
one_x['cap-color'] = encoder.transform(one_x['cap-color'])
one_x['cap-color']
```

```python
one_x['cap-color'].value_counts()
```
