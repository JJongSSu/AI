```python
import keras
print(keras.__version__)
```

```python
import numpy as np
```

```python
# 수집과정
x_train = np.array([0,1])
y_train = x_train*2 + 1
x_test = np.array([2,3])
y_test = x_test*2 + 1
# 학습모델 생성
model = keras.models.Sequential()
model.add(keras.layers.Dense(1, input_shape=(1,)))
model.summary()
# 한번 test 해봄
y_predict = model.predict(x_test)
print(y_predict.flatten())
print(y_test)
# 모델 픽스, 학습
model.compile('SGD', 'mse')
model.fit(x_train, y_train, epochs=1000, verbose=1)
# verbose = 0은 학습하는 동안 중간에 log를 출력하지 않는다는 뜻
y_predict = model.predict(x_test)
print(y_predict.flatten())
print(y_test)
```

단순 functional 모델

```python
x_train = np.array([0,1])
y_train = x_train*2 + 1
x_test = np.array([2,3])
y_test = x_test*2 + 1

x = keras.layers.Input( shape=(1,) ) # Dense, Tensor
d = keras.layers.Dense(1) 
y = d(x)

model = keras.models.Model(x,y)
model.summary()
y_predict = model.predict( x_test )
print( y_predict.flatten() )
print( y_test )
model.compile( 'SGD', 'mse' )
model.fit( x_train, y_train, epochs = 1000, verbose = 0 )
y_predict = model.predict( x_test ) 
print( y_predict.flatten() ) 
print( y_test )
```
